```python
import pandas 
```

```python
Revenue_Growth_Factor= 0.18
EBITDA_Margin= 0.45
Taxes= 0.21
Capital_Expenditures= 0.03
Working_Capital_Increase= 0.6
EBITDA_Exit_Multiple=32.70
Discount_Factor= 0.94
Discount_rate= 0.10
```

```python
start_year='2022'
end_year='2032'
```

```python
dates= pandas.date_range(start = '2022', end= '2032', freq = 'y')
dates
```

```python
df= pandas.DataFrame(index=dates)
df.loc['2022-12-31','Revenue']= 222* (1+Revenue_Growth_Factor )
df
```

```python
inital_revenue= 222
revenue= [inital_revenue*Revenue_Growth_Factor]
revenue
for i in range(1,len(dates)):
    revenue.append(revenue[i-1] * Revenue_Growth_Factor )

df['revenue']=revenue
df
```

```python
for i in range (1, len(df)):
    df.iloc[i]['Revenue']= df.iloc[i-1]['Revenue']* (1+ Revenue_Growth_Factor)
df
```

```python

df['EBITDA']= df['Revenue']*EBITDA_Margin
df['Taxes']= df['EBITDA']* -Taxes
df['Capital Expenditures']= df['Revenue']* -Capital_Expenditures
df['Working Capital']= df['Revenue']*-Working_Capital_Increase
#df['Cash Flow']= df['EBITDA']+df['Taxes']+df['Capital Expenditures']+df['Working Capital']
df['Cash Flow']=df.drop(columns='Revenue').sum(axis=1)

df
```

```python
df['Discount Factor']=[Discount_Factor ** i for i in range(1,len(dates) +1) ]
df
```

```python
df["Present Value Free Cash Flow"] = df["Cash Flow"] * df["Discount Factor"]
df
```

```python
print(f"""
Our projected Present Value of Projected Future Cash Flows are: ${df['Present Value Free Cash Flow'].sum().round(0)}.
We are providing revenue guidance in 2032 to be: ${round(df.iloc[-1]["Revenue"] * Revenue_Growth_Factor)}.
With an estimated EBITDA around: ${EBITDA_Exit_Multiple * EBITDA_Margin }.""")

```
